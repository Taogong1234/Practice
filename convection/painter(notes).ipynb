```python
# 两条线最终形态
import matplotlib.pyplot as plt
from matplotlib import animation 
import numpy as np

data1 = np.loadtxt('/home/taogang/desk/CFDtask/outfx.dat')
data2 = np.loadtxt('/home/taogang/desk/CFDtask/outgx.dat')
fig,ax=plt.subplots(figsize=(10,5),dpi=100)
xdata=np.linspace(0,15,151)
ydata1=data1[0,:]
ydata2=data2[0,:]
line1,=ax.plot(xdata,ydata1,lw=1, label='Computational')
line2,=ax.plot(xdata,ydata2,lw=3, label='Analytical')

def init():
    plt.xlim(-1,16)
    plt.ylim(-0.2,1.2)
    plt.xlabel('x')
    plt.ylabel('f(x,t)')
    plt.xticks(np.linspace(0,15,16))
    plt.yticks(np.linspace(0,1,11))
    plt.legend()
    plt.title('Convection')
    return line1,line2

def upgrade(i):
    line1.set_ydata(data1[i,:])
    line2.set_ydata(data2[i,:])
    return line1,line2
    
ani=animation.FuncAnimation(fig=fig,func=upgrade,frames=151,init_func=init,interval=50,blit=False)
# ani.save('666.gif',writer='pillow')
ani.save('555.mp4')
plt.show()
```

```python
# 动画例子1
import matplotlib.pyplot as plt
from matplotlib import animation
import numpy as np

fig, ax = plt.subplots()
xdata = []
ydata = []

line, = plt.plot(xdata, ydata, 'r-')

def init():
    ax.set_xlim(-1, 6)
    ax.set_ylim(-1, 30)
    return line,

def update(frame):  # 帧
    xdata.append(frame)
    ydata.append(frame ** 2)
    line.set_data(xdata, ydata)
    return line,

ani = animation.FuncAnimation(
    fig=fig,
    func=update,
    frames=np.linspace(0, 5, 100),
    init_func=init,
    interval=100,  # 时间间隔: 毫秒
    repeat=False)


# ani.save('2.gif', writer='pillow')
ani.save('666.mp4')# conda install -c conda-forge/label/broken ffmpeg
plt.show()
```
